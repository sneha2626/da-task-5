Checked missing values using isnull().sum().

```python
#Identify Missing Values
df.isnull().sum()
```

Filled numeric missing values using mean and categorical using mode.

```python
#Handle Missing Values (Cleaning)
#Numeric columns → mean / median
df['math score'] = df['math score'].fillna(df['math score'].mean())
```

```python
#Categorical columns → mode
# Check and handle missing values in gender column
mode_gender = df['gender'].mode()
if df['gender'].isnull().sum() > 0 and not mode_gender.empty:
    df['gender'] = df['gender'].fillna(mode_gender.iloc[0])
```

```python
df.columns
```

```python
#Remove Duplicates , Check duplicates
df.duplicated().sum()
```

```python
#Row count before
df.shape
```

```python
#Remove duplicates
df = df.drop_duplicates()
```

```python
#Row count after
df.shape
```

```python
#Datatype Conversion, Example: Convert categorical to category type (memory efficient)
df['gender'] = df['gender'].astype('category')
```

```python
#Create New Columns
#Example 1: Average Score
df['average_score'] = (
    df['math score'] +
    df['reading score'] +
    df['writing score']
) / 3

#Example 2: Performance Category

def performance(avg):
    if avg >= 75:
        return 'High'
    elif avg >= 50:
        return 'Medium'
    else:
        return 'Low'

df['performance_level'] = df['average_score'].apply(performance)
```

```python
#Save Cleaned Dataset
df.to_csv("cleaned_data.csv", index=False)
```
